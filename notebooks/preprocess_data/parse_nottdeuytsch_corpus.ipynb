```python
import pyreadr

result = pyreadr.read_r("../external_data/NottDeuYTSch_Corpus.rda")
```

```python
# Save each object as a separate parquet file
for obj_name, df in result.items():
    df.to_parquet(f"../external_data/{obj_name}.parquet", index=False)
    print(f"Saved {obj_name}.parquet")
```
